```python
import hashlib
from itertools import product
#x = "0009000012"
#h_obj = hashlib.sha256(str(x).encode()).hexdigest()

h_obj=input("hash a comprobar si esta en el rango: ")
for comb in product('0123456789', repeat=10):
    x = ''.join(comb)
    h = hashlib.sha256(str(x).encode()).hexdigest()
    if h == h_obj:
        print(f"numero encontrado", x)
        encontrado = True
        break
if not encontrado:
    print("numero no encontrado")
```

```python
class MerkleTree:    #hecho por gpt

    def __init__(self, transacciones):
        self.transacciones = transacciones
        self.root = self.construir_arbol()

    def hash(self, data):
        return hashlib.sha256(data.encode()).hexdigest()

    def construir_arbol(self):
        nivel = [self.hash(tx) for tx in self.transacciones]
        nuevos = []

        while len(nivel) > 1:
            i = 0
            while i < len(nivel):

                if i + 1 < len(nivel):
                    nuevo_hash = self.hash(nivel[i] + nivel[i + 1])
                    nuevos.append(nuevo_hash)
                else:
                    nuevos.append(nivel[i])
                i += 2
            nivel = nuevos
            nuevos = []

        return nivel[0]
```

```python
from itertools import permutations


encontrado = False
tx = []
i = input("numero de transacciones")
for i in range (0,i,1)
    x = input("transaccion")
    tx.append(x)
objetivo = input("ROOT_HASH_AQUI: ")

for perm in permutations(tx):
    arbol = MerkleTree(perm)
    root = arbol.root

    if root == objetivo:
        print("Orden correcto:", perm)
        encontrado = True
        break
if not encontrado:
    print("el root no existe con las transacciones dadas")
```
