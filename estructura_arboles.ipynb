```python
class NodoEstudiante:
    def __init__(self, estudiante):
        self.estudiante = lista_estudiantes # El diccionario con id, nombre, promedio
        self.id = lista_estudiantes["id"]   # Extraemos el ID para comparar rápido
        self.izquierda = None
        self.derecha = None

class ArbolBinario:
    def __init__(self):
        self.raiz = None

    # Método para insertar cuando la lista está desordenada
    def insertar_desordenado(self, estudiante):
        if self.raiz is None:
            self.raiz = NodoEstudiante(estudiante)
        else:
            self._insertar_recursivo(self.raiz, estudiante)

    def _insertar_recursivo(self, nodo_actual, estudiante):
        # Si el ID es menor, vamos a la izquierda
        if estudiante["id"] < nodo_actual.id:
            if nodo_actual.izquierda is None:
                nodo_actual.izquierda = NodoEstudiante(estudiante)
            else:
                self._insertar_recursivo(nodo_actual.izquierda, estudiante)
        # Si el ID es mayor, vamos a la derecha
        elif estudiante["id"] > nodo_actual.id:
            if nodo_actual.derecha is None:
                nodo_actual.derecha = NodoEstudiante(estudiante)
            else:
                self._insertar_recursivo(nodo_actual.derecha, estudiante)
    def buscar(self, id_buscado):
        """Busca un ID específico descartando la mitad del árbol en cada paso."""
        nodo_actual = self.raiz

        while nodo_actual is not None:

            # ¡Lo encontramos!
            if id_buscado == nodo_actual.id:
                #estudiante = nodo_actual.estudiante
                #print(f"es el estudiante {estudiante['nombre']}con promedio{estudiante['promedio']}")
                return nodo_actual.estudiante

            # Si lo que buscamos es MENOR, descartamos toda la derecha y vamos a la izquierda
            elif id_buscado < nodo_actual.id:
                nodo_actual = nodo_actual.izquierda

            # Si lo que buscamos es MAYOR, descartamos toda la izquierda y vamos a la derecha
            else:
                nodo_actual = nodo_actual.derecha

        #print(f"Estudiante con ID {id_buscado} no encontrado.")
        return None

mi_arbol = ArbolBinario()
```

```python
import math

class NodoBPlus:
    def __init__(self, es_hoja=False):
        self.es_hoja = es_hoja
        self.claves = []      # Guarda los IDs
        self.valores = []     # Si es hoja: guarda el diccionario del estudiante. Si es interno: guarda los nodos hijos.
        self.siguiente = None # El "gancho" del tren que conecta las hojas entre sí

class ArbolBPlus:
    def __init__(self, orden=4):
        # El 'orden' define cuántas claves máximas caben en un nodo
        self.raiz = NodoBPlus(es_hoja=True)
        self.orden = orden

    # --- 1. FUNCIÓN DE BÚSQUEDA ---
    def buscar(self, id_buscado):
        """Navega por las señales de tráfico hasta llegar a la hoja, luego busca el ID."""
        nodo_actual = self.raiz

        # 1. Bajar por el árbol hasta llegar a una hoja
        while not nodo_actual.es_hoja:
            encontrado = False
            for i, clave in enumerate(nodo_actual.claves):
                if id_buscado < clave:
                    nodo_actual = nodo_actual.valores[i] # Bajamos por el hijo correspondiente
                    encontrado = True
                    break
            # Si el ID es mayor que todas las claves, bajamos por el último hijo
            if not encontrado:
                nodo_actual = nodo_actual.valores[-1]

        # 2. Una vez en la hoja, buscamos el diccionario del estudiante
        for i, clave in enumerate(nodo_actual.claves):
            if clave == id_buscado:
                return nodo_actual.valores[i] # ¡Encontrado! Retorna el diccionario

        return None # No existe

    # --- 2. FUNCIÓN DE INSERCIÓN (Simplificada) ---
    def insertar(self, estudiante):
        """Inserta un estudiante y divide los nodos si se llenan (split)."""
        id_nuevo = lista_estudiantes["id"]
        raiz = self.raiz

        # Si la raíz se llena, necesitamos crear una nueva raíz por encima
        if len(raiz.claves) == self.orden - 1:
            nueva_raiz = NodoBPlus()
            self.raiz = nueva_raiz
            nueva_raiz.valores.append(raiz)
            self._dividir_hijo(nueva_raiz, 0)
            self._insertar_no_lleno(nueva_raiz, lista_estudiantes)
        else:
            self._insertar_no_lleno(raiz, lista_estudiantes)

    def _insertar_no_lleno(self, nodo, estudiante):
        id_nuevo = lista_estudiantes["id"]
        i = len(nodo.claves) - 1

        if nodo.es_hoja:
            # Encontrar la posición correcta e insertar clave y diccionario
            nodo.claves.append(0)
            nodo.valores.append(None)
            while i >= 0 and id_nuevo < nodo.claves[i]:
                nodo.claves[i + 1] = nodo.claves[i]
                nodo.valores[i + 1] = nodo.valores[i]
                i -= 1
            nodo.claves[i + 1] = id_nuevo
            nodo.valores[i + 1] = lista_estudiantes
        else:
            # Buscar el hijo correcto donde debe ir
            while i >= 0 and id_nuevo < nodo.claves[i]:
                i -= 1
            i += 1
            if len(nodo.valores[i].claves) == self.orden - 1:
                self._dividir_hijo(nodo, i)
                if id_nuevo > nodo.claves[i]:
                    i += 1
            self._insertar_no_lleno(nodo.valores[i], lista_estudiantes)

    def _dividir_hijo(self, nodo_padre, indice_hijo):
        """Magia del Árbol B+: Si un nodo se llena, se parte a la mitad."""
        orden = self.orden
        hijo_lleno = nodo_padre.valores[indice_hijo]
        nuevo_nodo = NodoBPlus(es_hoja=hijo_lleno.es_hoja)

        # Encontrar el punto medio
        medio = math.ceil(orden / 2) - 1

        # Subir la clave central al padre
        clave_sube = hijo_lleno.claves[medio]
        nodo_padre.claves.insert(indice_hijo, clave_sube)
        nodo_padre.valores.insert(indice_hijo + 1, nuevo_nodo)

        # Repartir las claves y valores
        nuevo_nodo.claves = hijo_lleno.claves[medio + 1:]
        hijo_lleno.claves = hijo_lleno.claves[:medio]

        if hijo_lleno.es_hoja:
            nuevo_nodo.valores = hijo_lleno.valores[medio + 1:]
            hijo_lleno.valores = hijo_lleno.valores[:medio + 1] # En hojas, la clave sube pero se mantiene abajo
            # Conectar las hojas (el tren)
            nuevo_nodo.siguiente = hijo_lleno.siguiente
            hijo_lleno.siguiente = nuevo_nodo
        else:
            nuevo_nodo.valores = hijo_lleno.valores[medio + 1:]
            hijo_lleno.valores = hijo_lleno.valores[:medio + 1]

mi_arbol_b = ArbolBPlus()
```

```python
import time
import random

def comparar_tiempos(lista_datos, abb, arbol_b, n_busquedas=1000):
    print(f"Iniciando prueba de rendimiento con {n_busquedas} búsquedas aleatorias...\n")

    # Extraemos n IDs aleatorios que SABEMOS que existen en 'lista_estudiantes'
    ids_a_buscar = random.sample([e["id"] for e in lista_datos], n_busquedas)

    # --- 1. PRUEBA DE LA LISTA (Fuerza Bruta) ---
    inicio_lista = time.perf_counter()
    for id_buscado in ids_a_buscar:
        for estudiante in lista_datos:
            if estudiante["id"] == id_buscado:
                break # Lo encontró, pasa al siguiente ID
    fin_lista = time.perf_counter()
    tiempo_lista = fin_lista - inicio_lista

    # --- 2. PRUEBA DEL ÁRBOL ABB ---
    inicio_abb = time.perf_counter()
    for id_buscado in ids_a_buscar:
        abb.buscar(id_buscado)
    fin_abb = time.perf_counter()
    tiempo_abb = fin_abb - inicio_abb

    # --- 3. PRUEBA DEL ÁRBOL B+ ---
    inicio_b = time.perf_counter()
    for id_buscado in ids_a_buscar:
        arbol_b.buscar(id_buscado)
    fin_b = time.perf_counter()
    tiempo_b = fin_b - inicio_b

    # --- RESULTADOS ---
    print("-" * 40)
    print("RESULTADOS DE LA PRUEBA")
    print("-" * 40)
    print(f"Lista clásica: {tiempo_lista:.5f} segundos")
    print(f"Árbol ABB:     {tiempo_abb:.5f} segundos")
    print(f"Árbol B+:      {tiempo_b:.5f} segundos")
    print("-" * 40)

    # Calcular cuántas veces más rápidos son los árboles comparados con la lista
    # Usamos max() para evitar dividir por cero si el árbol es instantáneo
    multiplicador_abb = tiempo_lista / max(tiempo_abb, 0.000001)
    multiplicador_b = tiempo_lista / max(tiempo_b, 0.000001)

    print(f"El ABB fue aprox. {multiplicador_abb:.0f} veces más rápido que la Lista.")
    print(f"El B+ fue aprox. {multiplicador_b:.0f} veces más rápido que la Lista.")

# --- EJECUTAR LA PRUEBA ---
# Aquí le pasamos tu variable exacta 'lista_estudiantes'
print("-----------------------------lista no ordenada------------------------------------------")
comparar_tiempos(lista_estudiantes, mi_arbol, mi_arbol_b, n_busquedas=100)
comparar_tiempos(lista_estudiantes, mi_arbol, mi_arbol_b, n_busquedas=1000)
comparar_tiempos(lista_estudiantes, mi_arbol, mi_arbol_b, n_busquedas=2000)
comparar_tiempos(lista_estudiantes, mi_arbol, mi_arbol_b, n_busquedas=3000)

print("----------------------------- lista ordenada  ------------------------------------------")
comparar_tiempos(lista_ord_estudiantes, mi_arbol, mi_arbol_b, n_busquedas=100)
comparar_tiempos(lista_ord_estudiantes, mi_arbol, mi_arbol_b, n_busquedas=1000)
comparar_tiempos(lista_ord_estudiantes, mi_arbol, mi_arbol_b, n_busquedas=2000)
comparar_tiempos(lista_ord_estudiantes, mi_arbol, mi_arbol_b, n_busquedas=3000)
```
